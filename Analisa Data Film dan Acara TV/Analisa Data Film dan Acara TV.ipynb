```python
# Menampilkan jumlah film dan acara TV dalam dataset
num_movies = df[df['type'] == 'Movie'].shape[0]
num_tv_shows = df[df['type'] == 'TV Show'].shape[0]
print(f"Jumlah Film: {num_movies}")
print(f"Jumlah Acara TV: {num_tv_shows}")
```

```python
# Menggunakan .str.replace() untuk menghilangkan "min" dari kolom 'Duration'
data = df['duration'] = df['duration'].str.replace('min', '').str.strip()
data = df['duration'] = df['duration'].str.replace('Season', '').str.strip()
data = df['duration'] = df['duration'].str.replace('Seasons', '').str.strip()
data
```

```python
# Menghapus teks "min" dari kolom 'Duration' dan mengubah menjadi tipe data numerik (integer)
df['duration'] = df['duration'].str.replace('min', '').astype(int)

# Menghitung rata-rata durasi film dari kolom 'Duration'
average_duration = df['duration'].mean()

print("Rata-rata durasi film:", average_duration, "min")
```

```python
# Mengurutkan DataFrame berdasarkan kolom "Rating" secara menurun (descending)
sorted_df = df.sort_values(by='rating', ascending=False)

# Mengambil 10 film dengan rating tertinggi
top_10_movies = sorted_df.head(10)
top_10_movies['title']
```

```python
# Menampilkan hanya data TV Shows
tv_shows_data = df[df['type'] == 'TV Show']

# Mengurutkan DataFrame berdasarkan kolom "Duration" secara menurun (descending)
sorted_tv_shows = tv_shows_data.sort_values(by='duration', ascending=False)

# Mengambil 10 acara TV dengan durasi terpanjang
top_10_tv_shows = sorted_tv_shows.head(10)

top_10_tv_shows
```

```python
# Menghitung frekuensi masing-masing genre
genre_counts = df['listed_in'].str.split(', ').explode().value_counts()

# Mengambil top 3 genre yang paling sering muncul
top_3_genres = genre_counts.nlargest(5)

print("Top 3 genre paling sering muncul:")
top_3_genres
```
